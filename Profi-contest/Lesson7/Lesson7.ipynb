# Подготовка данных

```python
# объединим все признаки в один датафрейм
X_train_full = pd.concat([formula_train, train], axis=1)
X_train_full = X_train_full.drop(['material', 'critical_temperature', 'critical_temp'], axis=1)

print(X_train_full.head())
print(X_train_full.shape)
```

```python
# выделим вектор ответов для тренировочных данных
y_train_full = train['critical_temperature']
print(y_train_full.shape)
```

```python
# проведем нормализацию данных
# StandartScaler() - стандартизация
# from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train_full)
X_train_full_norm = scaler.transform(X_train_full)
print(X_train_full_norm)
```

```python
# проделаем аналогичные действия для тестовых данных
X_full_test = pd.concat([formula_test, test], axis=1)
X_full_test = X_full_test.drop(['material'], axis=1)
print(X_full_test.shape)

# scaler = MinMaxScaler(feature_range=(0, 1))
# scaler.fit(X_full_test)
X_full_test_norm = scaler.transform(X_full_test)
print(X_full_test_norm)
```

```python
from xgboost import XGBRegressor

X_train, X_test, y_train, y_test = train_test_split(X_train_full_norm, y_train_full, test_size=0.3, random_state=1)

xgb = XGBRegressor()
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)

print('R_2_score:', r2_score(y_test, xgb_pred))
print('mean_absolute_error:', mean_absolute_error(y_test, xgb_pred))
print('mean_squared_error:', mean_squared_error(y_test, xgb_pred))
# 86%
```

```python
from sklearn.model_selection import KFold, GridSearchCV

kf = KFold(n_splits=5, shuffle=True, random_state=42)

xgb = XGBRegressor()

grid = {'n_estimators': [100, 300, 500], 'max_depth': [3, 5]}

xgb_cv = GridSearchCV(xgb, param_grid = grid, cv = kf, scoring='r2')
xgb_cv.fit(X_train_full_norm, y_train_full)

C = xgb_cv.best_params_
print("Best params:", C)
# > 15 минут
```

```python
# проверим какая оценка получилась при лучших параметрах
X_train, X_test, y_train, y_test = train_test_split(X_train_full_norm, y_train_full, test_size=0.3, random_state=1)

xgb = XGBRegressor(max_depth = 5, n_estimators = 500)
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)

print('R_2_score:', r2_score(y_test, xgb_pred))
print('mean_absolute_error:', mean_absolute_error(y_test, xgb_pred))
print('mean_squared_error:', mean_squared_error(y_test, xgb_pred))
# 92%
```
