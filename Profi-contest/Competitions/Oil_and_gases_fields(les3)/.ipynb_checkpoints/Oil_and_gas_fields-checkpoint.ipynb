# Сlassification of oil and gas fields

A serious problem arising in the early stages of development of oil and gas fields is the insufficient amount of information about the parameters of these fields. Incorrectly estimated parameters threaten with large errors in assessing the underlying volumes of oil and gas, which brings economic losses to the company that develops the field.

A particularly important parameter is the location of the oil field.

P.s.: This challenge and data was presented at the I-Professional Olympiad in 2021.

## Подготовка данных

### 1. Обработка категориальных признаков

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# преобразуем целевую переменную

labelencoder = LabelEncoder()
train['Onshore/Offshore'] = labelencoder.fit_transform(train['Onshore/Offshore'])

train.head()
```

```python
# преобразуем составные признаки

train_full = train.copy()

df1 = train_full['Tectonic regime'].str.get_dummies('/')
df1 = df1.add_prefix('Regime_').reset_index()
df1.drop(['index'], axis=1, inplace=True)

df2 = train_full['Structural setting'].str.get_dummies('/')
df2 = df2.add_prefix('Structural setting_').reset_index()
df2.drop(['index'], axis=1, inplace=True)

train_full = pd.concat([train, df1, df2], axis=1)
train_full
```

```python
# закодируем остальные категориальные признаки 

columns = ['Hydrocarbon type', 'Reservoir status', 'Period', 'Lithology']
train_full = pd.get_dummies(train_full, columns=columns)

train_full.head()
```

```python
# удалим ненужные столбцы после обработки
train_full.drop(['Tectonic regime', 'Structural setting'], axis=1, inplace=True)
train_full.head()
```

### 2. Балансировка классов целевой переменной

```python
from imblearn.over_sampling import SMOTE

# создадим модель SMOTE указав кол-во соседей 3 (больше нет смысла, у нас всего 5 экземпляров одного класса)
os = SMOTE(random_state=0, k_neighbors=3)

# вектор признаков
X_train_full = train_full.drop(['Onshore/Offshore'], axis=1) 
# вектор целевой переменной
y_train_full = train_full['Onshore/Offshore']

column = X_train_full.columns

print("До балансировки")
print(X_train_full.shape)
print(y_train_full.value_counts())


# Применим алгоритм балансировки
os_data_X, os_data_y = os.fit_resample(X_train_full, y_train_full)
os_data_X = pd.DataFrame(data=os_data_X, columns=column)
os_data_y = pd.DataFrame(data=os_data_y, columns=['Onshore/Offshore'])

print('_'*100)
print("После балансировки")
print(os_data_X.shape)
print(os_data_y.value_counts())
```

```python
train_os = pd.concat([os_data_X, os_data_y], axis=1)
train_os.head()
```

### 3. Корреляция признаков

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# data - это данные без кодировки категориальных признаков

cols = list(train_os.columns)
cols.pop(len(cols)-1)

fig, ax = plt.subplots(figsize=(10,8))
hm = sns.heatmap(train_os[cols].corr(), cbar=True, annot=False, ax=ax)
```

```python
cols = list(train_os.columns)
cols.pop(len(cols)-1)

train_os[cols].corr()
```

```python
# Feature Extraction with RFE
from sklearn.feature_selection import RFE
from catboost import CatBoostClassifier

cols = list(train_os.drop(columns=['Onshore/Offshore']).columns)
# print(cols)

X = train_os[cols].values
Y = train_os['Onshore/Offshore'].tolist()

# feature extraction

model = CatBoostClassifier()
rfe = RFE(model)
fit = rfe.fit(X, Y)
```

```python
type(fit)
```

```python
print("Num Features:", fit.n_features_)
print("Selected Features:", fit.support_)
print("Feature Ranking:", fit.ranking_)

print('*' * 100)
print(len(cols), len(fit.support_), len(fit.ranking_))
```

```python
df_rfe = pd.DataFrame({'feature': cols, 'rfe_label': fit.support_, 'rfe_score': fit.ranking_})
df_true = df_rfe[df_rfe['rfe_score']==True]

cols = list(df_true['feature'].values)

X = train_os[cols]
Y = train_os['Onshore/Offshore']
```

```python
X
```

```python
Y
```

### 4. Масштабирование признаков

```python
# применим масштабирование признаков 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_standard = scaler.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_standard, Y, test_size=0.3)
```

```python
X_train
```

```python
Y_train
```

### 5. Выбор модели

```python
# подключим необходимые библиотеки

# импортируем необходимые классификаторы
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb
from catboost import CatBoostClassifier

# имортируем метрики для оценки классифкаторов
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score

# Создадим модели классификаторов с дефолтным набором гиперпараметров
log = LogisticRegression()
tree = DecisionTreeClassifier()
svc = SVC()
knn = KNeighborsClassifier()
xgb = xgb.XGBClassifier()
catboost = CatBoostClassifier()
```

```python
from sklearn.model_selection import train_test_split

models = [log, tree, svc, knn, xgb, catboost]

# кроссвалидация, cross_val_score
```

```python
# Обучение каждой модели и получения

accuracy = []
recall = []
precision = []

for model in models:
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)

    accuracy.append(accuracy_score(Y_test, y_pred))
    recall.append(recall_score(Y_test, y_pred, average='macro'))
    precision.append(precision_score(Y_test, y_pred, average='macro'))
```

```python
df_score = pd.DataFrame({'models': models, 'accuracy': accuracy, 'recall': recall, 'precision': precision})
df_score
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# LogisticRegression

parameters = {'penalty': ['l1', 'l2', 'elasticnet', 'none'], 'dual': (True, False), 
              'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}

log = LogisticRegression()

log_clf = GridSearchCV(log, parameters)
log_clf.fit(X_train, Y_train)
```

```python
# DecisionTreeClassifier

parameters = {'criterion': ('gini', 'entropy'), 'min_samples_split':[1, 5], 'min_samples_leaf': [1, 5]}

tree = DecisionTreeClassifier()

tree_clf = GridSearchCV(tree, parameters)
tree_clf.fit(X_train, Y_train)
```

```python
# SVC
parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10], 'degree': [1, 5]}

svc = SVC()

svc_clf = GridSearchCV(svc, parameters)
svc_clf.fit(X_train, Y_train)
```

```python
# KNeighborsClassifier

parameters = {'n_neighbors': [1, 5], 'metric': ['euclidean', 'manhattan', 'chebyshev', 'minkowski'], }

knn = KNeighborsClassifier()

knn_clf = GridSearchCV(knn, parameters)
knn_clf.fit(X_train, Y_train)
```

```python
models = [log_clf, tree_clf, svc_clf, knn_clf, xgb, catboost]

# кроссвалидация, cross_val_score

# Обучение каждой модели и получения

accuracy = []
recall = []
precision = []

for model in models:
    y_pred = model.predict(X_test)

    accuracy.append(accuracy_score(Y_test, y_pred))
    recall.append(recall_score(Y_test, y_pred, average='macro'))
    precision.append(precision_score(Y_test, y_pred, average='macro'))
```

```python
df_score = pd.DataFrame({'models': models, 'accuracy': accuracy, 'recall': recall, 'precision': precision})
df_score
```
